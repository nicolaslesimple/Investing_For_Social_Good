```python
import numpy as np 
import requests

def get_symbol(symbol):
    url = "http://d.yimg.com/autoc.finance.yahoo.com/autoc?query={}&region=1&lang=en".format(symbol)

    result = requests.get(url).json()

    for x in result['ResultSet']['Result']:
        if x['symbol'] == symbol:
            return x['name']
```

```python
import pandas as pd

symbols = []
company = []
sector = []
industry = []
headquarters = []

for item in all_stocks: 
    symbols.append(item["symbol"] )
    company.append (item['company'])
    sector.append(item['sector'])
    industry.append(item['industry'])
    headquarters.append(item['headquarters'])
    
data = {'symbol' : symbols, 'company' : company, 'sector':sector, 'industry': industry, 'headquarters':headquarters}
df = pd.DataFrame(data)
```

```python
df_detailled = pd.DataFrame(df.headquarters.str.split(', ',1).tolist(), columns = ['city','state'])
df_detailled = df.assign(city = df_detailled.city, state = df_detailled.state)
df_detailled.drop(columns = 'headquarters', inplace = True )
df_detailled
```

```python
us_state_abbrev = {
    'AL': 'Alabama',
    'AK': 'Alaska',
    'AZ': 'Arizona',
    'AR': 'Arkansas',
    'CA': 'California',
    'CO': 'Colorado',
    'CT': 'Connecticut',
    'DE': 'Delaware',
    'FL': 'Florida',
    'GA': 'Georgia',
    'HI': 'Hawaii',
    'ID': 'Idaho',
    'IL': 'Illinois',
    'IN': 'Indiana',
    'IA': 'Iowa',
    'KS': 'Kansas',
    'KY': 'Kentucky',
    'LA': 'Louisiana',
    'ME': 'Maine',
    'MD': 'Maryland',
    'MA': 'Massachusetts',
    'MI': 'Michigan',
    'MN': 'Minnesota',
    'MS': 'Mississippi',
    'MO': 'Missouri',
    'MT': 'Montana',
    'NE': 'Nebraska',
    'NV': 'Nevada',
    'NH': 'New Hampshire',
    'NJ': 'New Jersey',
    'NM': 'New Mexico',
    'NY': 'New York',
    'NC': 'North Carolina',
    'ND': 'North Dakota',
    'OH': 'Ohio',
    'OK': 'Oklahoma',
    'OR': 'Oregon',
    'PA': 'Pennsylvania',
    'RI': 'Rhode Island',
    'SC': 'South Carolina',
    'SD': 'South Dakota',
    'TN': 'Tennessee',
    'TX': 'Texas',
    'UT': 'Utah',
    'VT': 'Vermont',
    'VA': 'Virginia',
    'WA': 'Washington',
    'WV': 'West Virginia',
    'WI': 'Wisconsin',
    'WY': 'Wyoming',
    'DC': 'D.C.', 
    'MP':'Northern Mariana Islands', 
    'PW' : 'Palau',  
    'PR':'Puerto Rico', 
    'VI': 'Virgin Islands',
}
```

```python
def convert_to_code(df_detailled):
    code = []
    noisy_data = []

    for i in df_detailled['state']: 
        found = False 
        for key, value in us_state_abbrev.items():
            if i.lower().strip() == value.lower().strip(): 
                code.append(str(key)) 
                found = True

        if not found: 
            noisy_data.append(i)

    #remove the noisy data 
    for i in noisy_data: 
        df_detailled = df_detailled[df_detailled['state'] != i]

    return df_detailled.assign(code = code)
```

```python
df_detailled = convert_to_code(df_detailled)
df_detailled.to_csv('localization_of_companies.csv')
```

```python
import plotly.plotly as py
import pandas as pd

df = df_mean
```

```python
for col in df.columns:
    df[col] = df[col].astype(str)

scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]


df['text'] = df['state'] + '<br>' + 'Price '+df['PRICE']+ '<br>' + ' Rank Overall ' + df['rank.overall']+ '<br>'+\
    'pcent'+ df['pcent'] + '<br>' +' USD ' + df['usd']

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df['code'],
        z = df['weighted_score'].astype(float),
        locationmode = 'USA-states',
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "[-]")
        ) ]

layout = dict(
        title = '2017 Social Investment Score by State',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```
